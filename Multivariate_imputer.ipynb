# Multivariate Imputer

```python
import pandas as pd
file = "https://centric-airbyte-demo-250087949428.s3.eu-north-1.amazonaws.com/data/airbyte_certification_distribution/2024_06_19_1718820651178_0.csv"
df = pd.read_csv(file)
```

### Create missing values

```python
import numpy as np
np.random.seed(seed=10)
h = np.random.choice(a=df.index, replace=False, size=30)
df.loc[h, 'salary_expectation'] = np.nan

```

```python
np.random.seed(seed=10)
h = np.random.choice(a=df.index, replace=False, size=20)
df.loc[h, 'height'] = np.nan
```

```python
df.isna().sum()
```

```python
numerical_columns = [col for col in df.columns if df[col].dtypes != "O"]
num_df = df[numerical_columns]
num_df
```

#### KNNImputer

```python
from sklearn.impute import KNNImputer
knn = KNNImputer(n_neighbors=5, add_indicator=True)
knn.fit(df[numerical_columns])
pd.DataFrame(knn.transform(df[numerical_columns]))
```

#### Mice Imputer (Iterative)

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp = IterativeImputer(imputation_order='roman')
pd.DataFrame(imp.fit_transform(num_df))
```
